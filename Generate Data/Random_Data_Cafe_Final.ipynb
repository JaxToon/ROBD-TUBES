```python
from faker import Faker
import random
import json
from datetime import date
import pandas as pd

# Inisialisasi faker
fake = Faker()

# Generate 10 data karyawan
data = []
product_id_map = {
    "Americano": "C1",
    "Cappuccino": "C2",
    "Latte": "C3",
    "Espresso": "C4",
    "Mocha": "C5",
    "Flat White": "C6",
    "Macchiato": "C7",
    "Cold Brew": "C8",
    "Matcha Latte": "NC1",
    "Chocolate Frappe": "NC2"
}
list_menu = ["Americano","Cappuccino","Latte","Espresso","Mocha","Flat White","Macchiato","Cold Brew","Matcha Latte","Chocolate Frappe"]
for _ in range(1098000):
    employee = {
        "id_transaction":(_+1),
        "product" : [],
        "name": fake.name()
    }

    order_quantity = 0
    curr_menu = []
    for i in range(random.randint(1, 5)):
        menu = random.choice(list_menu)
        if menu in curr_menu:
          i -= 1
          continue
        curr_menu.append(menu)
        id_product = product_id_map[menu]
        quantity = random.randint(1, 3)
        employee["product"].append({
            "id_product": id_product,
            "name": menu,
            "quantity": quantity
        })
        order_quantity += quantity
    employee["order_quantity"] = order_quantity
    data.append(employee)

# Simpan ke file JSON
with open("data_karyawan.json", "w") as f:
    json.dump(data, f, indent=4)

print("Data berhasil disimpan dalam file 'data_karyawan.json'")
```

```python
# Load the JSON data from the file
with open("data_karyawan.json", "r") as f:
    data = json.load(f)

# Create a DataFrame from the loaded data
df = pd.DataFrame(data)

# Display the DataFrame
df
```

```python
# Tambahkan kolom "id_franchise" dengan nilai 1 hingga 10 secara berurutan
df['id_franchise'] = [(i % 10) + 1 for i in range(len(df))]

# Tampilkan DataFrame dengan kolom baru
df
```

```python
# Fungsi untuk menghasilkan nilai random integer berdasarkan rentang id_franchise
def generate_employee_id(id_franchise):
  return random.randint((id_franchise-1)*5+1, (id_franchise-1)*5+5)


# Terapkan fungsi ke kolom 'id_franchise' untuk membuat kolom 'id_employee'
df['id_employee'] = df['id_franchise'].apply(generate_employee_id)

# Tampilkan DataFrame dengan kolom 'id_employee'
df
```

```python
df['transaction_date'] = pd.date_range(start='2024-01-01', end='2025-01-01', periods=len(df)).date
df
```

```python
df['transaction_date'] = df['transaction_date'].replace(date(2025, 1, 1), date(2024, 12, 31))
df
```

```python
df['transaction_date'] = pd.to_datetime(df['transaction_date'])
df.info()
```

```python
# Ubah kolom 'transaction_date' menjadi string dengan format YYYY-MM-DD
df['transaction_date'] = df['transaction_date'].dt.strftime('%Y-%m-%d')

# Simpan DataFrame ke file JSON
df.to_json("df_cafe.json", orient='records', indent=4)
```
